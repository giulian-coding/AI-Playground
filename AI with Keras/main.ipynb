```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
x = np.linspace(-10, 10, 100)
relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, relu, label='ReLU')
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.title('Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dropout


# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Preprocess data
X_train = X_train.reshape(-1, 784).astype('float32') / 255  # Flatten and normalize
X_test = X_test.reshape(-1, 784).astype('float32') / 255
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the model
# Adding Dropout to prevent overfitting
model = Sequential([
    Dense(128, input_shape=(784,), activation='relu'),
    Dropout(0.3),  # Dropout layer with 30% rate
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])


# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
# or
# Compile the model with different optimizers and loss functions
# model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])  # Using SGD (Stochastic Gradient Descent)

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_split=0.1)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc:.4f}')
```
